### Bar graph 

- vertical

```python
# Creating a plot with dictionary

book_prices = {
                "5 Am Club":250,
                "Atomic Habits":300,
                "Deep Work":240
}

fig,ax = plt.subplots()
ax.bar(book_prices.keys(),book_prices.values())
ax.set(title="Book Prices",ylabel = "cost(Rs)");
```

```python
# Horizontal bar graph

fig,ax = plt.subplots()
ax.barh(list(book_prices.keys()),list(book_prices.values()))
ax.set(title="Book Prices",xlabel="Cost(Rs)");
```

### Subplots
* Multiple plots on one figure
https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
# Option 1 : creating multiple subplots

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,
                                         ncols=2,
                                         figsize =(10,5)
                                        )

#plotting data

ax1.plot(x,x/2);
ax2.scatter(x,np.cos(x));
ax3.bar(book_prices.keys(),book_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2: creating multiple subplots

fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5)
                     )
# indexing should be done,indexing here is similar to matrix

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(100),np.random.random(100));
ax[1,0].bar(book_prices.keys(),book_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Plotting using Object Oriented Method

- https://matplotlib.org/matplotblog/posts/pyplot-vs-object-oriented-interface/
- Obiect Oriented method can be used for more complicated methods

```python
heart_disease.head()
```

```python
# performing data analysis on patients over 50+ age

over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method

over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
            figsize=(10,6));
```

```python
## OO Method mixed with pyplot

fig, ax = plt.subplots(figsize=(10,6))

over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
             ax=ax);
ax.set_xlim([45,100]);
```

```python
## OO Method from scratch

# create a plot
fig,ax = plt.subplots(figsize=(10,6))

# plot the data

scatter = ax.scatter(over_50["age"],
                     over_50["chol"],
                     c=over_50["target"])


# customize the plot

ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

ax.legend(*scatter.legend_elements(),title="Target")

# add the mean line

ax.axhline(over_50["chol"].mean(),
          linestyle="--");

```

## Adding Another plot to existing styled one




                            

```python
# setup plot

fig,(ax0,ax1) = plt.subplots(
                            nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True
                            );

# Add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]
                     )

# customize ax0

ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="Target")

# setup a mean line

ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--",
           color="b",
           label="Average");

# add data for ax1
scatter = ax1.scatter(x=heart_disease["age"],
                      y=heart_disease["thalach"], # thalach = heart beat rate
                      c=heart_disease["target"])

# customize ax1

ax1.set(title="Heart Disease and Max Heart Rate Levels",
        ylabel="Max Heart rate",
       xlabel="Age")
ax1.legend(*scatter.legend_elements(),title="Target")

# setup mean line for ax1

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--",
            color="b",
            label="Average");

# Add title to fig

fig.suptitle("Heart Disease Data Analysis", fontsize=16 , fontweight ="bold");
        
```

## Changing xlim & ylim

```python
## Before the change (we've had color updates)

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
# after adding in different x and y limitations



fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 10))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# set a x_lim
ax0.set_xlim([50,80])

# Setup a mean line
ax0.axhline(y=over_50["chol"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axis 1, 1 (row 1, column 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Setup y_lim
ax1.set_ylim([60,200])


# Setup a mean line
ax1.axhline(y=over_50["thalach"].mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
ax1.legend(*scatter.legend_elements(), title="Target")

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
