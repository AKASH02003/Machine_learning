```python
# A matplotlib workflow

# 0.Import and ready matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. get data ready
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot

fig,ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title ="Sample plot",
        xlabel="x-axis",
        ylabel="y-axis")
# 5. save and show
fig.savefig("sampleplot.png")
```

# Making the most frequently used plots using Numpy arrays

The most common types of plots are


- `line`
- `scatter`
- `bar`
- `hist`
- `subplots()`

```python
import numpy as np
```

### Line plot

- Line plot is a default visualization type in matplotlib.

```python
# create an numpy array

x = np.linspace(0,10,100)
x[:10]
```

```python
# plotting

fig,ax =plt.subplots()
ax.plot(x,x**2)
```

### Scatter plot


```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig,ax= plt.subplots()
ax.scatter(x,np.sin(x));
```

### Bar graph 

- vertical
- horizontal



```python
# Creating a plot with dictionary

book_prices = {
                "5 Am Club":250,
                "Atomic Habits":300,
                "Deep Work":240
}

fig,ax = plt.subplots()
ax.bar(book_prices.keys(),book_prices.values())
ax.set(title="Book Prices",ylabel = "cost(Rs)");
```

```python
# Horizontal bar graph

fig,ax = plt.subplots()
ax.barh(list(book_prices.keys()),list(book_prices.values()))
ax.set(title="Book Prices",xlabel="Cost(Rs)");
```

### Histogram (hist)



```python
# Making some data for normal distribution

x = np.random.randn(1000) # pulls random normal distribution

# plotting

fig,ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.random(1000) # random sitribution

fig,ax = plt.subplots()
ax.hist(x)
```

### Subplots
* Multiple plots on one figure
https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
# Option 1 : creating multiple subplots

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,
                                         ncols=2,
                                         figsize =(10,5)
                                        )

#plotting data

ax1.plot(x,x/2);
ax2.scatter(x,np.cos(x));
ax3.bar(book_prices.keys(),book_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2: creating multiple subplots

fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5)
                     )
# indexing should be done,indexing here is similar to matrix

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(100),np.random.random(100));
ax[1,0].bar(book_prices.keys(),book_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

# Working with actual data

let's do a little data manipulation on our `car_sales` DataFrame

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '',regex =True)
car_sales
```

```python
#Removing last two zeros

car_sales["Price"]= car_sales["Price"].str[:-2]
car_sales
```

```python
# Add a date column

car_sales["Sale Date"] = pd.date_range("1/1/2023",periods=len(car_sales))
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.dtypes
```

```python
# Make total sales column
#car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x= "Sale Date",y="Total Sales");
```
